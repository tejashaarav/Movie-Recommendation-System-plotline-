```python
## As a cutoff, we'll choose the 95th percentile. To put it another way, a film must receive more votes than at least 95% of the other films on the list in order to be included in the rankings.
```

```python
m= movies_data['vote_count'].quantile(0.90)
m
```

```python
q_df = movies_data[movies_data['vote_count']>=m]
```

```python
q_df
```

```python
C= q_df['vote_average'].mean()
C
```

```python
def imdb_score(x):
    C= 6.9629937629937615
    m = 1838.4000000000015
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m) * r) + (m/(m+v) * C)
```

```python
q_df['imdb-weight-score'] = q_df.apply(lambda x : imdb_score(x),axis = 1)
```

```python
###We'll create a function that takes a movie title as an input and returns a list of the top ten most comparable films. To begin, a reverse mapping of movie names and DataFrame indices is required. To put it another way, we need a way to get the index of a movie in our metadata DataFrame based on its title.
```

```python
movie_titles = pd.Series(df_plotline.index, index=df_plotline['title']).drop_duplicates()
```

```python
movie_titles
```

```python
def get_recs(title,cosine_similarity = cosine_similarity):
    title = movie_titles[title]
    sim_scores = list(enumerate(cosine_similarity[title]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:16]
    movie_indices = [i[0] for i in sim_scores]
    return df_plotline['title'].iloc[movie_indices]
```
